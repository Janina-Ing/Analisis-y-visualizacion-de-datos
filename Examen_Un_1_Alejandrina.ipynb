# IMPORTAR LIBRERIAS

```python
import pandas as pd
```

# CARGAS EL ARCHIVO EXCEL

```python
df = pd.read_excel('examen_transacciones.xlsx')
```

# VISUALIZAR LOS PRIMEROS 10 REGISTROS

```python
print("Primeros 10 registros:")
print(df.head(10))
```

# IMPRIMIR NOMBRE DE LAS VARIABLES

```python
print("nColumnas del Data Frame")
print(df.columns)
```

# CREAR COLUMNA "Subtotal"

```python
df["Subtotal"] = df["Cantidad"]*df["Precio_Unitario"]
```

CREAR COLUMNA "IVA Trasladado"
# Insertamos a la derecha de Subtotal

```python
pos = df.columns.get_loc('Subtotal') + 1
df.insert(pos, 'IVA Trasladado', df['Subtotal'] * 0.16)
```

#FILTRAR TRANSACCIONES COMPLETADAS

```python
df['Subtotal'] = df['Cantidad'] * df['Precio_Unitario']
```

# CALCULAR TOTAL DE VENTAS POR PRODUCTO (CON IVA)

```python
df_completadas = df[df['Estado'] == 'Completado'].copy()
df_completadas["Total_con_IVA"] = df_completadas["Subtotal"] + df_completadas["IVA Trasladado"]
ventas_producto = df_completadas.groupby("Producto")["Total_con_IVA"].sum().reset_index()
print("\nVentas por producto (con IVA):")
print(ventas_producto)
```

# TOTAL DE VENTAS (CON IVA)

```python
Total_Ventas = df_completadas['Total_con_IVA'].sum()
print("\nTotal de Ventas con IVA:", Total_Ventas)
```

# M√ÅXIMO DE VENTAS POR CLIENTE

```python
ventas_cliente = df_completadas.groupby('Cliente_ID')['Total_con_IVA'].sum()
df_completadas['Total_ventas_cliente'] = df_completadas['Cliente_ID'].map(ventas_cliente)
```

# CREAR COLUMNA "Max_Ventas_cliente"

```python
df_completadas['Max_ventas_cliente'] = df_completadas['Cliente_ID'].map(ventas_cliente)
```

# PROMEDIO DE VENTAS POR PORDUCTO

```python
mean_ventas_producto = df_completadas.groupby('Producto')['Total_con_IVA'].mean()
print("\nPromedio de ventas por producto:")
print(mean_ventas_producto)
```

# CREAR COLUMNA "Mean_ventas_producto"

```python
df_completadas['Mean_ventas_producto'] = df_completadas['Producto'].map(mean_ventas_producto)
```

# GUARDAR RESUMEN EN EXCEL

```python
ventas_producto.to_excel('/content/Ventas_completadas_ELAO.xlsx', index=False)
print("\nArchivo guardado como Ventas_completadas_ELAO.xlsx")
```
